#### Imports

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.colors as colors
import seaborn as sns
import itertools
from scipy.stats import norm
import scipy.stats
from sklearn.naive_bayes import GaussianNB

%matplotlib inline
sns.set()
```

#### Data Generation

```python
#data of Gaussian Distributions
np.random.seed(seed = 3)
#distribution1
mu1 = [2, 2]
sigma1 = [[5, -4], [-4, 5]]
x1, y1 = np.random.multivariate_normal(mu1, sigma1, 100).T
#distribution2
mu2 = [-2, -2]
sigma2 = [[3.5, 3], [3, 3.5]]
x2, y2 = np.random.multivariate_normal(mu2, sigma2, 100).T
```

#### Excercise: A) Plot the decision boundary

```python
""" Code for comparing the decision boundary for the optimal and the naive Bayes classifiers goes here """

def optimal_bayes_class(X, mu_list, sigma_list): 
    
    '''This funtion returns the optimal class given X'''
    max_score = -1000000;
    max_score_class = -1
    classes = len(mu_list)
    for clas in range(classes):
        #Computing the posterior probability for each class
        score = scipy.stats.multivariate_normal.pdf(X, mean=mu_list[clas], cov=sigma_list[clas])
        if(score > max_score):
            max_score = score
            max_score_class = clas    
    return max_score_class

# Defining plot parameters and added x1,y1 and x2,y2 scatter plots

fig = plt.figure(figsize = (20,9))
subplot = fig.add_subplot(1, 2, 1)
subplot.set_title('Optimal Bayes Classifier (Green) vs Naive Bayes Classifier (Red)')
subplot.set_ylabel('Y axis')
subplot.set_xlabel('X axis')
subplot.scatter(x1, y1, label ="Gaussian 1")
subplot.scatter(x2, y2, label ="Gaussian 2")

# Defining parameters to be used

mu_list = [mu1,mu2]
sigma_list = [sigma1,sigma2]
X = np.linspace(-8, 8, 100)
Y = np.linspace(-8, 8, 100)
X, Y = np.meshgrid(X, Y)
X_data = np.asarray(np.vstack((np.hstack((x1,x2)),np.hstack((y1,y2)))).T)
y_labels = np.hstack((np.zeros(100),np.ones(100)))

# Training Naive Bayes model

model_sk = GaussianNB(priors = None)
model_sk.fit(X_data,y_labels)

# ********************  For Optimal Bayes Classification *****************

pred = []
for a,b in zip(np.ravel(X),np.ravel(Y)):
    value = optimal_bayes_class(np.array([a,b]),mu_list,sigma_list)
    pred.append(value)
pred = np.array(pred)
Z1 = pred.reshape(X.shape)

# ******************** For Naive Bayes Classification ********************

pred = []
for a,b in zip(np.ravel(X),np.ravel(Y)):
    value = model_sk.predict([[a,b]])[0]
    pred.append(value)
pred = np.array(pred)
Z2 = pred.reshape(X.shape)

# **************** Plotting boundary lines ************************

subplot.contour(X, Y, Z1, 1, colors = "green")
subplot.contourf(X, Y, Z1, 1, alpha=0.1, colors = ["blue", "darkred"])
subplot.contour(X, Y, Z2, 1, colors = 'red')
plt.legend(loc="best")
plt.show()
```

#### Excercise: B) Accuracies

```python
""" Code for comparing the accuracies of the optimal and the naive Bayes classifiers goes here """
print()
# Optimal Bayes accuracy

y_pred = []
for a, b in zip(np.ravel(X_data[:,0]), np.ravel(X_data[:,1])):
    value = optimal_bayes_class(np.array([a,b]),mu_list,sigma_list) 
    y_pred.append(value)
y_pred = np.array(y_pred)
optimal_accuracy = np.mean(y_pred == y_labels)
print("Optimal Bayes Accuracy is:", optimal_accuracy)

# Naive Bayes accuracy

naive_accuracy = model_sk.score(X_data,y_labels)
print("Naive Bayes Accuracy is:", naive_accuracy)
```
