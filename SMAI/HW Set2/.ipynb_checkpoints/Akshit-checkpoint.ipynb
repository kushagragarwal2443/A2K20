# Optimal Bayes Classifier - Excercise Notebook


### Comparing Optimal Bayes and Naive Bayes using simulated Gaussian data for non-zero covariance.

#### Imports

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.colors as colors
import seaborn as sns
import itertools
from scipy.stats import norm
import scipy.stats
from sklearn.naive_bayes import GaussianNB

%matplotlib inline
sns.set()
```

#### Data Generation

```python
#data of Gaussian Distributions
np.random.seed(seed = 3)

#distribution1
mu1 = [2, 2]
sigma1 = [[5, -4], [-4, 5]]
x1, y1 = np.random.multivariate_normal(mu1, sigma1, 100).T

#distribution2
mu2 = [-2, -2]
sigma2 = [[3.5, 3], [3, 3.5]]
x2, y2 = np.random.multivariate_normal(mu2, sigma2, 100).T
```

#### Excercise: A) Plot the decision boundary

```python
#%%%%%%%%%%%%%% Excercise: Write the code below %%%%%%%%%%%%%%%%%%%%%%%%%
"""
Code for comparing the decision boundary for the optimal and the naive Bayes classifiers goes here
"""
def optimal_class(co_ord,mu_list,sigma_list): 
    scores_list = []
    
    for p in range(len(sigma_list)):
        score = scipy.stats.multivariate_normal.pdf(co_ord, mean=mu_list[p], cov=sigma_list[p])
        scores_list.append(score)
             
    return np.argmax(scores_list)

plt.rcParams["figure.figsize"] = (6,6)
plt.rcParams.update({'font.size': 20})

mu_list = [mu1,mu2]
sigma_list = [sigma1,sigma2]

X = np.linspace(-8, 8, 100)
Y = np.linspace(-8, 8, 100)
X, Y = np.meshgrid(X, Y)

#Optimal Bayes Classifier
z=[]
for x, y in zip(np.ravel(X), np.ravel(Y)) :
    z.append(optimal_class(np.array([x,y]),mu_list,sigma_list))

z = np.array(z)
Z = z.reshape(X.shape)

plt.contour( X, Y, Z, 1,colors="green")

#Naive Bayes Classifier
x_train = np.asarray(np.vstack((np.hstack((x1,x2)),np.hstack((y1,y2)))).T)
y_train = np.hstack((np.zeros(100),np.ones(100)))

naive_model = GaussianNB(priors = None)
naive_model.fit(x_train,y_train)

z = []
for x, y in zip(np.ravel(X), np.ravel(Y)):
    z.append(naive_model.predict( [[x,y]])[0])
z = np.array(z)
Z = z.reshape(X.shape)
plt.contour( X, Y, Z, 1, colors = ('red'))
plt.scatter(x1, y1, alpha = 0.7,label="distribution 1")
plt.scatter(x2, y2, alpha = 0.7,label="distribution 2")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Optimal(green) vs Naive(red) Bayes')
plt.legend()
plt.show()
```

#### Excercise: B) Accuracies

```python
#%%%%%%%%%%%%%% EXCERCISE - Write the code below %%%%%%%%%%%%%%%%%%%%%%%%%%%
"""
Code for comparing the accuracies of the optimal and the naive Bayes classifiers goes here
"""
#Optimal Bayes accuracy
y_pred=[]
for xx, yy in zip(np.ravel(x_train[:,0]), np.ravel(x_train[:,1])):
    y_pred.append(optimal_class(np.array([xx,yy]),mu_list,sigma_list))

y_pred = np.array(y_pred)
print("Optimal Bayes accuracy:",np.mean(y_pred == y_train))

#Naive Bayes accuracy
print("Naive Bayes accuracy:",naive_model.score(x_train,y_train))
```
