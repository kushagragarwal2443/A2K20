```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def calc_gaussian(x, mean, std):
    return math.exp(-1 * ((float(x)-float(mean))**2) / (2*(float(std)**2)) ) / (math.sqrt(2*math.pi)*float(std))

calc_gaussian(7,5,5)
```

```python
mean1 = 3.0
mean2 = 5.0
stdev = 0.5
index_array = []
bigarray1 = []
bigarray2 = []
for x in np.arange(0,8,0.01):
    value1 = calc_gaussian(x, mean1, stdev)
    value2 = calc_gaussian(x, mean2, stdev)
    bigarray1.append(value1)
    bigarray2.append(value2)
    index_array.append(x)
```

```python
plt.plot(index_array, bigarray1, label="Mean = 3")
plt.plot(index_array, bigarray2, label="Mean = 5")
plt.legend(loc="best")
plt.show()
```
