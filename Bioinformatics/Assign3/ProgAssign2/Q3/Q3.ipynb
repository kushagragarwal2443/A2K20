### I have used the genome for the bacteria: Staphylococcus aureus subsp. aureus NCTC 8325

```python
file = open("sequence.fasta")
sequence=""
for line in file:
    if(line[0]!=">"):
        sequence+=line[:-1]
print(len(sequence))
```

```python
gc_content=[] # to store G+C content
at_dev=[] # to store deviation form A=T
gc_dev=[] # to store deviation form G=C
gc_total=0 # compute total percentage of GC in the genome

window = 50000 # Window of 50kb used

for i in range(0,len(sequence),window):
    
    region = sequence[i:i+window]
    
    # To store individual bases counts in the window
    count_A,count_G, count_T, count_C = 0,0,0,0
    
    for j in range(len(region)):
        
        if(region[j]=="A"):
            count_A+=1
        if(region[j]=="T"):
            count_T+=1
        if(region[j]=="G"):
            count_G+=1
            gc_total+=1
        if(region[j]=="C"):
            count_C+=1
            gc_total+=1
      
    # computing the various quantities
    val1 = (count_G + count_C)/(count_A+count_T+count_G+count_C)
    val2 = (count_A - count_T)/ (count_A + count_T)
    val3 = (count_G - count_C)/ (count_G + count_C)
    
    # Append to the respective arrays
    gc_content.append(val1)
    at_dev.append(val2)
    gc_dev.append(val3)
```

## Plotting the 3 values against the windows

```python
import matplotlib.pyplot as plt
import numpy as np
x_axis=[]
for i in range(len(gc_content)):
    x_axis.append(i+1)
plt.figure(figsize=(10,5))
plt.plot(x_axis, gc_content)
plt.plot(x_axis, at_dev)
plt.plot(x_axis, gc_dev)
plt.legend(["G+C content","AT skew", "GC skew" ],loc="best")
plt.show()
```

## Printing out the 3 arrays

```python
print("G+C Contents\n")
print(gc_content)
print("\nDeviation from AT\n")
print(at_dev)
print("\nDeviation from GC\n")
print(gc_dev)
```
