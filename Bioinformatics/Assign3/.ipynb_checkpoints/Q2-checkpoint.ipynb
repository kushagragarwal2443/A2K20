```python
def countlength(length):
    num_digits= 0
    while length>0:
        length = int(length/10)
        num_digits+=1
    return num_digits

def formword(a, i):
    final_word=""
    b = countlength(i)
    for k in range(a-b):
        final_word+=" "
    final_word+=str(i)
    
    return final_word

def fasta2gbk():
    
    fd1 = input("Enter Fasta file: ")
    fd2 = input("Enter Flat file Format: ")
#     fd1 = "sequence.fasta"
#     fd2 = "a.gbk"
    file = open(fd1, "r")
    sequence=""
    sequence_name=""
    for line in file:
        if(line[0]==">"):
            sequence_name = line[1:-1]
        else:
            line = line.lower()
            sequence+=line[:-1]
            
    length = len(sequence)
    
    file2 = open(fd2, "w")
    
    ID = sequence_name.split(" ")[0]
    desc = sequence_name[len(ID)+1: ]
    
    file2.write("LOCUS      "+'\t' + str(ID) + "\t" + str(length) + " bp\n")
    file2.write("DESCRIPTION"+"\t"+str(desc)+"\n")
    file2.write("\nORIGIN\n")
    
    string_size = countlength(length)
    
    for i in range(0,len(sequence),10):
        
        word = formword(string_size,i+1)
        if(i%60==0):
            file2.write("\n"+str(word)+" "+sequence[i:i+10]+" ")
        else:
            file2.write(sequence[i:i+10]+" ")
            
fasta2gbk()
```
